Model C: Rate of Tempo Change + Tempo Change Percentage

```python
import pandas as pd
import statsmodels.formula.api as smf

# Load your data
df = pd.read_csv("/content/Combined with Rate of Tempo Change, Tempo Change Percentage, Arousal Change, and Normalization.csv")

# === AROUSAL MODEL === #

# REML version
model_arousal_reml = smf.mixedlm(
    "Arousal ~ Normalized_Rate_of_Tempo_Change + Normalized_Tempo_Change_Percentage",
    data=df,
    groups=df["Song_x"]
)
result_arousal_reml = model_arousal_reml.fit()
print("\n--- Mixed Effects Model: AROUSAL (REML) ---")
print(result_arousal_reml.summary())

# REML Fit Stats
loglik_arousal_reml = result_arousal_reml.llf
reml_deviance_arousal = -2 * loglik_arousal_reml
print("\nREML Model Fit Stats (Arousal):")
print("Log-Likelihood:", loglik_arousal_reml)
print("REML Deviance:", reml_deviance_arousal)

# ML version
model_arousal_ml = smf.mixedlm(
    "Arousal ~ Normalized_Rate_of_Tempo_Change + Normalized_Tempo_Change_Percentage",
    data=df,
    groups=df["Song_x"]
)
result_arousal_ml = model_arousal_ml.fit(reml=False)
print("\n--- Mixed Effects Model: AROUSAL (ML) ---")
print(result_arousal_ml.summary())

# ML Fit Stats
loglik_arousal_ml = result_arousal_ml.llf
ml_deviance_arousal = -2 * loglik_arousal_ml
aic_arousal_ml = result_arousal_ml.aic
bic_arousal_ml = result_arousal_ml.bic
print("\nML Model Fit Stats (Arousal):")
print("Log-Likelihood:", loglik_arousal_ml)
print("ML Deviance:", ml_deviance_arousal)
print("AIC:", aic_arousal_ml)
print("BIC:", bic_arousal_ml)



```

```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro

# Load your data
df = pd.read_csv("/content/Combined with Rate of Tempo Change, Tempo Change Percentage, Arousal Change, and Normalization.csv")

# === AROUSAL MODEL (REML) === #
model_arousal_reml = smf.mixedlm(
    "Arousal ~ Normalized_Rate_of_Tempo_Change + Normalized_Tempo_Change_Percentage",
    data=df,
    groups=df["Song_x"]
)
result_arousal_reml = model_arousal_reml.fit()
print("\n--- Mixed Effects Model: AROUSAL (REML) ---")
print(result_arousal_reml.summary())

# Extract residuals and fitted values
residuals = result_arousal_reml.resid
fitted = result_arousal_reml.fittedvalues

# --- ASSUMPTION CHECKS ---

# 1. Q-Q plot (Normality of residuals)
sm.qqplot(residuals, line='s')
plt.title("Q-Q Plot of Residuals (Arousal Model)")
plt.show()

# 2. Histogram of residuals
sns.histplot(residuals, kde=True)
plt.title("Histogram of Residuals (Arousal Model)")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()

# 3. Residuals vs Fitted values (Homoscedasticity)
plt.scatter(fitted, residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residuals vs Fitted Values")
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.show()

# 4. Shapiro-Wilk Test for normality
stat, p = shapiro(residuals)
print(f"\nShapiro-Wilk Test for Normality: W={stat:.4f}, p-value={p:.4f}")
if p > 0.05:
    print("✅ Residuals likely follow normal distribution (fail to reject H0)")
else:
    print("❗ Residuals may not be normally distributed (reject H0)")

# 5. (Optional) Breusch-Pagan test – Not perfect for mixed models, but here for reference
from statsmodels.stats.diagnostic import het_breuschpagan
# Use the fixed effects exog matrix
exog = result_arousal_reml.model.exog
bp_test = het_breuschpagan(residuals, exog)
print(f"\nBreusch-Pagan Test: LM Stat={bp_test[0]:.4f}, p-value={bp_test[1]:.4f}")
```
